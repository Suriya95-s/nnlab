```python
import tensorflow as tf


scalar = tf.constant(7)
print("Scalar:", scalar)
print("The number of dimensions of the scalar is:", scalar.ndim)
vector = tf.constant([10, 10])
print("Vector:", vector)
print("The number of dimensions of the vector is:", vector.ndim)
matrix = tf.constant([[1, 2], [3, 4]])
print("Matrix:\n", matrix)
print("The number of dimensions of the matrix is:", matrix.ndim)
matrix = tf.constant([[1, 2], [3, 4]])
matrix1 = tf.constant([[2, 4], [6, 8]])

print("Addition of two matrices:")
print(matrix + matrix1)
```
